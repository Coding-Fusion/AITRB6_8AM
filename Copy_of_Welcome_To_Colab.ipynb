```
import yfinance as yf
import pandas as pd
import ta

def check_buy_signal_improved(stock_symbol, rsi_period=14, short_ma=50, long_ma=200, atr_period=14, target_multiplier=2, stop_multiplier=1):
    # Fetch 2 years of daily data
    # Fetching more data allows for longer-term indicators (like MA200) to fully form
    df = yf.download(stock_symbol, period="2y")

    # Drop any missing data
    df.dropna(inplace=True)

    # --- Calculate Indicators ---

    # Calculate RSI
    # .squeeze() is used here to convert the potentially single-column DataFrame/Series to a Series
    rsi_indicator = ta.momentum.RSIIndicator(df['Close'].squeeze(), window=rsi_period)
    df['RSI'] = rsi_indicator.rsi()

    # Moving Averages
    df['MA50'] = df['Close'].rolling(window=short_ma).mean()
    df['MA200'] = df['Close'].rolling(window=long_ma).mean()

    # ATR (Average True Range)
    # .squeeze() is used here to convert the potentially single-column DataFrame/Series to a Series
    atr_indicator = ta.volatility.AverageTrueRange(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=atr_period)
    df['ATR'] = atr_indicator.average_true_range()

    # MACD (Moving Average Convergence Divergence)
    # Adding MACD for trend confirmation or divergence signals
    macd_indicator = ta.trend.MACD(close=df['Close'].squeeze(), window_slow=26, window_fast=12, window_sign=9)
    df['MACD'] = macd_indicator.macd()
    df['MACD_Signal'] = macd_indicator.macd_signal()
    df['MACD_Diff'] = macd_indicator.macd_diff() # Histogram

    # Bollinger Bands
    # Adding Bollinger Bands for volatility and potential overbought/oversold areas relative to recent price
    bb_indicator = ta.volatility.BollingerBands(close=df['Close'].squeeze(), window=20, window_dev=2)
    df['BB_Lower'] = bb_indicator.bollinger_lband()
    df['BB_Upper'] = bb_indicator.bollinger_hband()
    df['BB_MA'] = bb_indicator.bollinger_mavg() # Middle band (usually 20-day SMA)

    # Drop any new NaNs after indicators
    # Dropping NaNs is crucial as indicators like MAs and MACD need a certain history
    df.dropna(inplace=True)

    # Check if there is enough data after dropping NaNs
    if df.empty:
        print(f"\nAnalyzing {stock_symbol}:")
        print("❌ Not enough data to calculate indicators after dropping NaNs.")
        return # Exit function if not enough data

    # Use the most recent row
    latest = df.iloc[-1]

    # Extract scalar values using .item()
    current_price = latest['Close'].item()
    atr_value = latest['ATR'].item()
    latest_rsi = latest['RSI'].item()
    latest_ma50 = latest['MA50'].item()
    latest_ma200 = latest['MA200'].item()
    latest_macd = latest['MACD'].item()
    latest_macd_signal = latest['MACD_Signal'].item()
    latest_macd_diff = latest['MACD_Diff'].item()
    latest_bb_lower = latest['BB_Lower'].item()
    latest_bb_upper = latest['BB_Upper'].item()


    # --- Define Improved Conditions ---

    # Condition 1: RSI buy signal (could be refined, e.g., RSI bouncing from below 30)
    rsi_buy_signal = latest_rsi < 30

    # Condition 2: Golden Cross (MA50 crossing above MA200) - your original logic
    # Check if there is a previous row to compare for crossover
    if len(df) >= 2:
        previous_ma50 = df['MA50'].iloc[-2].item()
        previous_ma200 = df['MA200'].iloc[-2].item()
        ma_crossover_signal = (latest_ma50 > latest_ma200 and previous_ma50 <= previous_ma200)
    else:
        ma_crossover_signal = False # Not enough data for crossover check

    # Condition 3: MACD Crossover (MACD line crossing above Signal line)
    # Check if there is a previous row for MACD crossover
    if len(df) >= 2:
         previous_macd = df['MACD'].iloc[-2].item()
         previous_macd_signal = df['MACD_Signal'].iloc[-2].item()
         macd_crossover_signal = (latest_macd > latest_macd_signal and previous_macd <= previous_macd_signal)
    else:
        macd_crossover_signal = False # Not enough data for MACD crossover check


    # Condition 4: Price is near the lower Bollinger Band
    bb_buy_signal = current_price <= latest_bb_lower

    # --- Combine Conditions for a Stronger Signal ---

    # A potential strategy: Require a combination of signals for a buy
    # Example: RSI oversold OR price near lower BB, AND a bullish MA or MACD signal
    # This makes the signal more stringent than just checking one condition.
    # You can adjust this logic based on your specific strategy.

    # Example Logic:
    # Buy if (RSI < 30 OR Price near Lower BB) AND (Golden Cross OR MACD Crossover)
    strong_buy_signal = (rsi_buy_signal or bb_buy_signal) and (ma_crossover_signal or macd_crossover_signal)

    # You could also define different signal levels (e.g., weak, medium, strong)
    # weak_signal = rsi_buy_signal or ma_crossover_signal or macd_crossover_signal or bb_buy_signal # Any single signal
    # medium_signal = (rsi_buy_signal or bb_buy_signal) and (ma_crossover_signal or macd_crossover_signal) # Combination as above
    # strong_signal = rsi_buy_signal and ma_crossover_signal and macd_crossover_signal # All three crossover signals (very strict)


    # --- Targets and Stop Loss (Using ATR) ---
    # These remain the same, calculated from the latest ATR

    target_price = current_price + (atr_value * target_multiplier)
    stop_loss_price = current_price - (atr_value * stop_multiplier)


    # --- Print Analysis ---
    print(f"\nAnalyzing {stock_symbol}:")
    print(f"📈 Current Price: ${current_price:.2f}")
    print(f"📊 RSI: {latest_rsi:.2f}")
    print(f"📉 MA50: {latest_ma50:.2f}, MA200: {latest_ma200:.2f}")
    print(f"📈 MACD: {latest_macd:.2f}, Signal: {latest_macd_signal:.2f}, Diff: {latest_macd_diff:.2f}")
    print(f"📊 BB Lower: {latest_bb_lower:.2f}, BB Upper: {latest_bb_upper:.2f}")
    print(f"📏 ATR (Volatility): {atr_value:.2f}")

    # Print status of individual potential signals
    if rsi_buy_signal:
        print("📉 RSI < 30 → Potential oversold condition.")
    if ma_crossover_signal:
         print("🟢 MA50 crossed above MA200 → Golden cross detected.")
    if macd_crossover_signal:
        print("🔵 MACD crossed above Signal line → Bullish momentum.")
    if bb_buy_signal:
         print("⬇️ Price near Lower Bollinger Band → Potential support/reversal.")


    # --- Final Buy Signal Check ---
    # Check the combined condition(s)

    if strong_buy_signal: # Using the example combined signal
        print("\n✅ STRONG Potential BUY Signal (Combination of indicators met)")
        print(f"🎯 Target Price: ${target_price:.2f}")
        print(f"🛑 Suggested Stop Loss: ${stop_loss_price:.2f}")
    # elif medium_signal: # Example of checking a medium signal
    #      print("\n☑️ MEDIUM Potential BUY Signal (Combination of indicators met)")
    #      print(f"🎯 Target Price: ${target_price:.2f}")
    #      print(f"🛑 Suggested Stop Loss: ${stop_loss_price:.2f}")
    # elif weak_signal: # Example of checking a weak signal
    #      print("\n⚪ WEAK Potential BUY Signal (At least one indicator met)")
    #      print(f"🎯 Target Price: ${target_price:.2f}")
    #      print(f"🛑 Suggested Stop Loss: ${stop_loss_price:.2f}")
    else:
        print("\n❌ No strong buy signal right now based on current criteria.") # Updated message


# Example usage with improved function
check_buy_signal_improved("GOOGL") # Replace with any stock ticker
# check_buy_signal_improved("MSFT")
# check_buy_signal_improved("GOOGL")
```

```
import yfinance as yf
import pandas as pd
import ta

def check_buy_signal_optimized(stock_symbol, rsi_period=14, short_ma=50, long_ma=200, atr_period=14, bb_window=20, bb_dev=2, macd_fast=12, macd_slow=26, macd_sign=9, target_multiplier=2, stop_multiplier=1):
    # Fetch 2 years of daily data (or more if needed for very long MAs)
    df = yf.download(stock_symbol, period="2y")

    # Drop any missing data
    df.dropna(inplace=True)

    # --- Calculate Indicators ---

    # Calculate RSI
    rsi_indicator = ta.momentum.RSIIndicator(df['Close'].squeeze(), window=rsi_period)
    df['RSI'] = rsi_indicator.rsi()

    # Moving Averages
    df['MA50'] = df['Close'].rolling(window=short_ma).mean()
    df['MA200'] = df['Close'].rolling(window=long_ma).mean()

    # ATR (Average True Range)
    atr_indicator = ta.volatility.AverageTrueRange(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=atr_period)
    df['ATR'] = atr_indicator.average_true_range()

    # MACD
    macd_indicator = ta.trend.MACD(close=df['Close'].squeeze(), window_slow=macd_slow, window_fast=macd_fast, window_sign=macd_sign)
    df['MACD'] = macd_indicator.macd()
    df['MACD_Signal'] = macd_indicator.macd_signal()
    df['MACD_Diff'] = macd_indicator.macd_diff() # Histogram

    # Bollinger Bands
    bb_indicator = ta.volatility.BollingerBands(close=df['Close'].squeeze(), window=bb_window, window_dev=bb_dev)
    df['BB_Lower'] = bb_indicator.bollinger_lband()
    df['BB_Upper'] = bb_indicator.bollinger_hband()
    df['BB_MA'] = bb_indicator.bollinger_mavg() # Middle band

    # Drop any new NaNs after indicators
    df.dropna(inplace=True)

    # Check if there is enough data after dropping NaNs (needs at least 2 rows for crossover checks)
    if len(df) < 2:
        print(f"\nAnalyzing {stock_symbol}:")
        print("❌ Not enough historical data to calculate indicators and crossovers after dropping NaNs.")
        return # Exit function if not enough data

    # Use the most recent row and the previous row
    latest = df.iloc[-1]
    previous = df.iloc[-2]


    # Extract scalar values using .item() for latest data
    current_price = latest['Close'].item()
    atr_value = latest['ATR'].item()
    latest_rsi = latest['RSI'].item()
    latest_ma50 = latest['MA50'].item()
    latest_ma200 = latest['MA200'].item()
    latest_macd = latest['MACD'].item()
    latest_macd_signal = latest['MACD_Signal'].item()
    latest_macd_diff = latest['MACD_Diff'].item()
    latest_bb_lower = latest['BB_Lower'].item()
    latest_bb_upper = latest['BB_Upper'].item()
    latest_bb_ma = latest['BB_MA'].item() # Middle band

    # Extract scalar values using .item() for previous day data (for crossover checks)
    previous_rsi = previous['RSI'].item()
    previous_ma50 = previous['MA50'].item()
    previous_ma200 = previous['MA200'].item()
    previous_macd = previous['MACD'].item()
    previous_macd_signal = previous['MACD_Signal'].item()
    previous_macd_diff = previous['MACD_Diff'].item()


    # --- Define Fine-Tuned Conditions ---

    # Condition 1: RSI Bullish Signal (RSI crossing up through oversold level)
    # More specific than just being below 30
    rsi_bullish_cross = (latest_rsi > 30 and previous_rsi <= 30)

    # Condition 2: Golden Cross (MA50 crossing above MA200) - your original logic
    golden_cross = (latest_ma50 > latest_ma200 and previous_ma50 <= previous_ma200)

    # Condition 3: Price above MA50 (indicates shorter-term uptrend)
    price_above_ma50 = current_price > latest_ma50

    # Condition 4: MACD Bullish Crossover (MACD line crossing above Signal line)
    macd_bullish_cross = (latest_macd > latest_macd_signal and previous_macd <= previous_macd_signal)

    # Condition 5: MACD Histogram turning positive (momentum shifting up)
    macd_hist_turning_up = (latest_macd_diff > 0 and previous_macd_diff <= 0)

    # Condition 6: Price is near or below the lower Bollinger Band and bouncing
    # Could look for price closing above the lower band after touching it
    bb_bounce = (current_price > latest_bb_lower and previous['Close'].item() <= latest_bb_lower)


    # --- Combine Conditions for Optimized Signals ---

    # Example of an "Optimized" Buy Signal
    # Requires multiple signals to align, suggesting confluence
    # Logic could be: (RSI confirms undervaluation/bounce OR BB bounce)
    # AND (MA trend is positive OR a bullish MA/MACD cross)
    # AND (MACD histogram confirms upward momentum)

    optimized_buy_signal = (
        (rsi_bullish_cross or bb_bounce) and  # Undervaluation/Bounce signal
        (latest_ma50 > latest_ma200 or golden_cross or macd_bullish_cross) and # Trend or Bullish Momentum signal
        (latest_macd_diff > 0) # MACD Histogram confirms positive momentum
    )

    # You can create different tiers of signals based on combinations
    # Another example: A simpler signal for more frequent entries
    # simpler_buy_signal = macd_bullish_cross and price_above_ma50

    # --- Targets and Stop Loss (Using ATR) ---
    target_price = current_price + (atr_value * target_multiplier)
    stop_loss_price = current_price - (atr_value * stop_multiplier)


    # --- Print Analysis ---
    print(f"\nAnalyzing {stock_symbol}:")
    print(f"📈 Current Price: ${current_price:.2f}")
    print(f"📊 RSI: {latest_rsi:.2f} (Previous: {previous_rsi:.2f})")
    print(f"📉 MA50: {latest_ma50:.2f} (Previous: {previous_ma50:.2f}), MA200: {latest_ma200:.2f} (Previous: {previous_ma200:.2f})")
    print(f"📈 MACD: {latest_macd:.2f} (Previous: {previous_macd:.2f}), Signal: {latest_macd_signal:.2f} (Previous: {previous_macd_signal:.2f}), Diff: {latest_macd_diff:.2f} (Previous: {previous_macd_diff:.2f})")
    print(f"📊 BB Lower: {latest_bb_lower:.2f}, BB Upper: {latest_bb_upper:.2f}, BB Middle: {latest_bb_ma:.2f}")
    print(f"📏 ATR (Volatility): {atr_value:.2f}")

    # Print status of individual potential signals
    if rsi_bullish_cross:
        print("📉 RSI crossed above 30 → Potential end of oversold.")
    if golden_cross:
         print("🟢 MA50 crossed above MA200 → Golden cross detected.")
    if price_above_ma50:
         print("Above MA50 → Shorter-term trend positive.")
    if macd_bullish_cross:
        print("🔵 MACD crossed above Signal line → Bullish momentum.")
    if macd_hist_turning_up:
        print("📊 MACD Histogram turning positive → Upward momentum increasing.")
    if bb_bounce:
        print("⬆️ Price bounced from Lower Bollinger Band.")


    # --- Final Buy Signal Check ---
    if optimized_buy_signal: # Using the example combined signal
        print("\n✅ OPTIMIZED Potential BUY Signal (Multiple conditions met)")
        print(f"🎯 Target Price: ${target_price:.2f}")
        print(f"🛑 Suggested Stop Loss: ${stop_loss_price:.2f}")
    # elif simpler_buy_signal: # Example of checking a simpler signal
    #      print("\n☑️ SIMPLER Potential BUY Signal (Alternative conditions met)")
    #      print(f"🎯 Target Price: ${target_price:.2f}")
    #      print(f"🛑 Suggested Stop Loss: ${stop_loss_price:.2f}")
    else:
        print("\n❌ No optimized buy signal right now based on current criteria.") # Updated message


# Example usage with optimized function
check_buy_signal_optimized("GOOGL") # Replace with any stock ticker
# check_buy_signal_optimized("MSFT")
# check_buy_signal_optimized("PLTR")
```

```
import yfinance as yf
import pandas as pd
import ta

def check_buy_signal_horizon(stock_symbol, short_rsi_period=14, short_ma=20, medium_ma=50, long_ma=200, atr_period=14, bb_window=20, bb_dev=2, macd_fast=12, macd_slow=26, macd_sign=9, target_multiplier=2, stop_multiplier=1):
    # Fetch enough data for all indicators, including the longest MA
    df = yf.download(stock_symbol, period="5y") # Fetch 5 years for robust long-term analysis

    # Drop any missing data
    df.dropna(inplace=True)

    # --- Calculate Indicators ---

    # Calculate RSI (can use same period for both horizons or separate)
    rsi_indicator = ta.momentum.RSIIndicator(df['Close'].squeeze(), window=short_rsi_period)
    df['RSI'] = rsi_indicator.rsi()

    # Moving Averages (using short, medium, and long MAs)
    df['MA_Short'] = df['Close'].rolling(window=short_ma).mean()
    df['MA_Medium'] = df['Close'].rolling(window=medium_ma).mean()
    df['MA_Long'] = df['Close'].rolling(window=long_ma).mean()

    # ATR
    atr_indicator = ta.volatility.AverageTrueRange(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=atr_period)
    df['ATR'] = atr_indicator.average_true_range()

    # MACD
    macd_indicator = ta.trend.MACD(close=df['Close'].squeeze(), window_slow=macd_slow, window_fast=macd_fast, window_sign=macd_sign)
    df['MACD'] = macd_indicator.macd()
    df['MACD_Signal'] = macd_indicator.macd_signal()
    df['MACD_Diff'] = macd_indicator.macd_diff()

    # Bollinger Bands
    bb_indicator = ta.volatility.BollingerBands(close=df['Close'].squeeze(), window=bb_window, window_dev=bb_dev)
    df['BB_Lower'] = bb_indicator.bollinger_lband()
    df['BB_Upper'] = bb_indicator.bollinger_hband()
    df['BB_MA'] = bb_indicator.bollinger_mavg()

    # Drop any new NaNs after indicators
    df.dropna(inplace=True)

    # Check if there is enough data
    if len(df) < max(short_ma, medium_ma, long_ma, bb_window, atr_period, macd_slow): # Ensure enough data for all indicators
        print(f"\nAnalyzing {stock_symbol}:")
        print("❌ Not enough historical data to calculate all indicators after dropping NaNs.")
        return # Exit function

    # Use the most recent row and the previous row
    latest = df.iloc[-1]
    previous = df.iloc[-2]

    # Extract scalar values from the latest row
    current_price = latest['Close'].item()
    atr_value = latest['ATR'].item()
    latest_rsi = latest['RSI'].item()
    latest_ma_short = latest['MA_Short'].item()
    latest_ma_medium = latest['MA_Medium'].item()
    latest_ma_long = latest['MA_Long'].item()
    latest_macd = latest['MACD'].item()
    latest_macd_signal = latest['MACD_Signal'].item()
    latest_macd_diff = latest['MACD_Diff'].item()
    latest_bb_lower = latest['BB_Lower'].item()
    latest_bb_upper = latest['BB_Upper'].item()
    latest_bb_ma = latest['BB_MA'].item()

    # Extract scalar values from the previous row (for crossover/change checks)
    previous_rsi = previous['RSI'].item()
    previous_ma_short = previous['MA_Short'].item()
    previous_ma_medium = previous['MA_Medium'].item()
    previous_ma_long = previous['MA_Long'].item()
    previous_macd = previous['MACD'].item()
    previous_macd_signal = previous['MACD_Signal'].item()
    previous_macd_diff = previous['MACD_Diff'].item()
    previous_close = previous['Close'].item()


    # --- Define Conditions for Short-Term Buy Signal ---
    # Short-term signals often look for quick momentum shifts or bounces

    # Short-term RSI bullish cross (e.g., crossing above 30 or 50)
    short_rsi_bullish_cross = (latest_rsi > 30 and previous_rsi <= 30) or \
                              (latest_rsi > 50 and previous_rsi <= 50) # Example: crossing above 50 for momentum confirmation

    # Price crossing above short-term MA
    price_cross_short_ma = (current_price > latest_ma_short and previous_close <= previous_ma_short)

    # Short-term MA cross (e.g., MA_Short crossing above MA_Medium)
    short_ma_cross = (latest_ma_short > latest_ma_medium and previous_ma_short <= previous_ma_medium)

    # MACD bullish cross (faster signals)
    short_macd_bullish_cross = (latest_macd > latest_macd_signal and previous_macd <= previous_macd_signal)

    # Price bouncing from Lower BB
    short_bb_bounce = (current_price > latest_bb_lower and previous_close <= latest_bb_lower)

    # Example Short-Term Buy Logic:
    # Require a momentum shift (RSI or MACD cross) AND price showing strength (above short MA or BB bounce)
    short_term_buy_signal = (short_rsi_bullish_cross or short_macd_bullish_cross) and \
                            (price_cross_short_ma or short_bb_bounce)

    # --- Define Conditions for Long-Term Buy Signal ---
    # Long-term signals often confirm major trend shifts or identify value in a long-term uptrend

    # Long-term trend confirmation (MA_Medium > MA_Long, generally bullish)
    long_term_uptrend = latest_ma_medium > latest_ma_long

    # Golden Cross (MA_Medium crossing above MA_Long)
    golden_cross = (latest_ma_medium > latest_ma_long and previous_ma_medium <= previous_ma_long)

    # Price is above long-term MA
    price_above_long_ma = current_price > latest_ma_long

    # MACD is above zero (confirms bullish momentum in the long-term context)
    macd_above_zero = latest_macd > 0

    # Example Long-Term Buy Logic:
    # Require long-term trend to be bullish (or just turning bullish) AND momentum confirmation
    long_term_buy_signal = (long_term_uptrend or golden_cross) and macd_above_zero and price_above_long_ma
    # Alternative: Just require a Golden Cross for entry
    # long_term_buy_signal = golden_cross


    # --- Targets and Stop Loss (Using ATR) ---
    # You might use different ATR multipliers for short vs long term,
    # but for simplicity, we'll calculate based on current ATR.
    short_term_target = current_price + (atr_value * target_multiplier)
    short_term_stop_loss = current_price - (atr_value * stop_multiplier)

    # Long-term targets/stops are often set differently, maybe based on
    # major support/resistance levels or larger ATR multiples.
    # Using a larger multiplier for long-term example:
    long_term_target = current_price + (atr_value * target_multiplier * 3) # Example: 3x the short-term target distance
    long_term_stop_loss = current_price - (atr_value * stop_multiplier * 2) # Example: Wider stop for long-term volatility


    # --- Print Analysis ---
    print(f"\nAnalyzing {stock_symbol} for different horizons:")
    print(f"📈 Current Price: ${current_price:.2f}")
    print(f"📊 RSI ({short_rsi_period}): {latest_rsi:.2f}")
    print(f"📉 MAs: Short({short_ma}): {latest_ma_short:.2f}, Medium({medium_ma}): {latest_ma_medium:.2f}, Long({long_ma}): {latest_ma_long:.2f}")
    print(f"📈 MACD: {latest_macd:.2f}, Signal: {latest_macd_signal:.2f}, Diff: {latest_macd_diff:.2f}")
    print(f"📊 BB Lower: {latest_bb_lower:.2f}, BB Upper: {latest_bb_upper:.2f}")
    print(f"📏 ATR ({atr_period}): {atr_value:.2f}")

    print("\n--- Potential Signals ---")

    # Print status of individual potential signals for clarity
    if short_rsi_bullish_cross: print("✔ Short-term RSI bullish cross detected.")
    if price_cross_short_ma: print("✔ Price crossed above short-term MA.")
    if short_ma_cross: print("✔ Short MA crossed above Medium MA.")
    if short_macd_bullish_cross: print("✔ Short-term MACD bullish cross detected.")
    if short_bb_bounce: print("✔ Price bounced from Lower Bollinger Band.")
    if long_term_uptrend: print("✔ Medium MA is above Long MA (Long-term uptrend).")
    if golden_cross: print("✔ Golden Cross (Medium MA above Long MA) detected.")
    if price_above_long_ma: print("✔ Price is above Long MA.")
    if macd_above_zero: print("✔ MACD is above zero (Bullish momentum).")


    # --- Final Buy Signal Check by Horizon ---
    print("\n--- Trading Horizon Signals ---")

    if short_term_buy_signal:
        print("✅ Potential SHORT-TERM BUY Signal")
        print(f"🎯 Suggested Short-Term Target: ${short_term_target:.2f}")
        print(f"🛑 Suggested Short-Term Stop Loss: ${short_term_stop_loss:.2f}")
    else:
        print("❌ No short-term buy signal right now.")

    if long_term_buy_signal:
        print("✅ Potential LONG-TERM BUY Signal")
        print(f"🎯 Suggested Long-Term Target: ${long_term_target:.2f}")
        print(f"🛑 Suggested Long-Term Stop Loss: ${long_term_stop_loss:.2f}")
    else:
        print("❌ No long-term buy signal right now.")


# Example usage with horizon-based signals
check_buy_signal_horizon("CRDO") # Example 1
# check_buy_signal_horizon("MSFT") # Example 2
# check_buy_signal_horizon("PLTR") # Example 3
# check_buy_signal_horizon("SPY") # Example 4 (ETF)
```

```
import yfinance as yf
import pandas as pd
import ta

def check_buy_signal_advanced(stock_symbol, short_rsi_period=14, short_ma=20, medium_ma=50, long_ma=200, atr_period=14, bb_window=20, bb_dev=2, macd_fast=12, macd_slow=26, macd_sign=9, stoch_k_period=14, stoch_d_period=3, adx_period=14, target_multiplier=2, stop_multiplier=1):
    # Fetch enough data for all indicators, including the longest MA and ADX/Stoch periods
    # Fetching 5 years should be sufficient for standard parameters
    df = yf.download(stock_symbol, period="5y")

    # Drop any missing data
    df.dropna(inplace=True)

    # --- Calculate Indicators ---

    # Calculate RSI
    rsi_indicator = ta.momentum.RSIIndicator(df['Close'].squeeze(), window=short_rsi_period)
    df['RSI'] = rsi_indicator.rsi()

    # Moving Averages
    df['MA_Short'] = df['Close'].rolling(window=short_ma).mean()
    df['MA_Medium'] = df['Close'].rolling(window=medium_ma).mean()
    df['MA_Long'] = df['Close'].rolling(window=long_ma).mean()

    # ATR
    atr_indicator = ta.volatility.AverageTrueRange(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=atr_period)
    df['ATR'] = atr_indicator.average_true_range()

    # MACD
    macd_indicator = ta.trend.MACD(close=df['Close'].squeeze(), window_slow=macd_slow, window_fast=macd_fast, window_sign=macd_sign)
    df['MACD'] = macd_indicator.macd()
    df['MACD_Signal'] = macd_indicator.macd_signal()
    df['MACD_Diff'] = macd_indicator.macd_diff()

    # Bollinger Bands
    bb_indicator = ta.volatility.BollingerBands(close=df['Close'].squeeze(), window=bb_window, window_dev=bb_dev)
    df['BB_Lower'] = bb_indicator.bollinger_lband()
    df['BB_Upper'] = bb_indicator.bollinger_hband()
    df['BB_MA'] = bb_indicator.bollinger_mavg()

    # Stochastic Oscillator
    stoch_indicator = ta.momentum.StochasticOscillator(high=df['High'], low=df['Low'], close=df['Close'], window=stoch_k_period, smooth_window=stoch_d_period)
    df['STOCH_K'] = stoch_indicator.stoch()
    df['STOCH_D'] = stoch_indicator.stoch_signal()

    # On-Balance Volume (OBV)
    obv_indicator = ta.volume.OnBalanceVolumeIndicator(close=df['Close'], volume=df['Volume'])
    df['OBV'] = obv_indicator.on_balance_volume()

    # Average Directional Index (ADX) - Note: ADX requires several periods to calculate properly
    adx_indicator = ta.trend.ADXIndicator(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=adx_period)
    df['ADX'] = adx_indicator.adx()
    df['ADX_POS'] = adx_indicator.adx_pos() # +DI
    df['ADX_NEG'] = adx_indicator.adx_neg() # -DI


    # Drop any new NaNs after indicators
    # ADX and OBV might introduce NaNs at the beginning
    df.dropna(inplace=True)

    # Check if there is enough data (needs at least 2 rows for crossover checks, and more for ADX/longest MA)
    required_data_points = max(short_ma, medium_ma, long_ma, bb_window, atr_period, macd_slow, stoch_k_period, stoch_d_period, adx_period * 2) # ADX often needs ~2*period
    if len(df) < required_data_points:
         print(f"\nAnalyzing {stock_symbol}:")
         print(f"❌ Not enough historical data to calculate all indicators after dropping NaNs. Need at least {required_data_points} data points.")
         return # Exit function


    # Use the most recent row and the previous row
    latest = df.iloc[-1]
    previous = df.iloc[-2]

    # Extract scalar values from the latest row
    current_price = latest['Close'].item()
    atr_value = latest['ATR'].item()
    latest_rsi = latest['RSI'].item()
    latest_ma_short = latest['MA_Short'].item()
    latest_ma_medium = latest['MA_Medium'].item()
    latest_ma_long = latest['MA_Long'].item()
    latest_macd = latest['MACD'].item()
    latest_macd_signal = latest['MACD_Signal'].item()
    latest_macd_diff = latest['MACD_Diff'].item()
    latest_bb_lower = latest['BB_Lower'].item()
    latest_bb_upper = latest['BB_Upper'].item()
    latest_bb_ma = latest['BB_MA'].item()
    latest_stoch_k = latest['STOCH_K'].item()
    latest_stoch_d = latest['STOCH_D'].item()
    latest_obv = latest['OBV'].item() # OBV is cumulative, absolute value less important than trend
    latest_adx = latest['ADX'].item()
    latest_adx_pos = latest['ADX_POS'].item()
    latest_adx_neg = latest['ADX_NEG'].item()

    # Extract scalar values from the previous row (for crossover/change checks)
    previous_rsi = previous['RSI'].item()
    previous_ma_short = previous['MA_Short'].item()
    previous_ma_medium = previous['MA_Medium'].item()
    previous_ma_long = previous['MA_Long'].item()
    previous_macd = previous['MACD'].item()
    previous_macd_signal = previous['MACD_Signal'].item()
    previous_macd_diff = previous['MACD_Diff'].item()
    previous_close = previous['Close'].item()
    previous_stoch_k = previous['STOCH_K'].item()
    previous_stoch_d = previous['STOCH_D'].item()
    previous_obv = previous['OBV'].item()
    previous_adx = previous['ADX'].item()
    previous_adx_pos = previous['ADX_POS'].item()
    previous_adx_neg = previous['ADX_NEG'].item()


    # --- Define Conditions for Short-Term Buy Signal (using more indicators) ---
    # Short-term signals often look for quick momentum shifts or bounces, supported by volume/trend strength

    # Short-term RSI bullish cross (crossing above 30 or 50)
    short_rsi_bullish_cross = (latest_rsi > 30 and previous_rsi <= 30) or \
                              (latest_rsi > 50 and previous_rsi <= 50)

    # Price crossing above short-term MA
    price_cross_short_ma = (current_price > latest_ma_short and previous_close <= latest_ma_short) # Use latest MA_Short for comparison

    # Short-term MA cross (e.g., MA_Short crossing above MA_Medium)
    short_ma_cross = (latest_ma_short > latest_ma_medium and previous_ma_short <= previous_ma_medium)

    # MACD bullish cross (faster signals)
    short_macd_bullish_cross = (latest_macd > latest_macd_signal and previous_macd <= previous_macd_signal)

    # Price bouncing from Lower BB
    short_bb_bounce = (current_price > latest_bb_lower and previous_close <= latest_bb_lower)

    # Stochastic Bullish Crossover (Stoch K crossing above Stoch D, typically below 20)
    stoch_bullish_cross_oversold = (latest_stoch_k > latest_stoch_d and previous_stoch_k <= previous_stoch_d and latest_stoch_k < 20)

    # OBV confirming upward movement (latest OBV is higher than previous OBV)
    obv_uptick = latest_obv > previous_obv

    # ADX confirming trend strength (ADX > 20-25) AND +DI > -DI (bullish direction)
    adx_confirm_bullish_momentum = (latest_adx > 20 and latest_adx_pos > latest_adx_neg)


    # Example Short-Term Buy Logic (more specific combination):
    # Require a quick momentum signal (Stoch or MACD cross)
    # AND Price showing strength (above short MA or BB bounce)
    # AND Volume confirms AND ADX confirms trend direction
    short_term_buy_signal = (stoch_bullish_cross_oversold or short_macd_bullish_cross) and \
                            (price_cross_short_ma or short_bb_bounce) and \
                            obv_uptick and \
                            adx_confirm_bullish_momentum

    # --- Define Conditions for Long-Term Buy Signal (using more indicators) ---
    # Long-term signals often confirm major trend shifts or identify value in a long-term uptrend

    # Long-term trend confirmation (MA_Medium > MA_Long, generally bullish)
    long_term_uptrend = latest_ma_medium > latest_ma_long

    # Golden Cross (MA_Medium crossing above MA_Long)
    golden_cross = (latest_ma_medium > latest_ma_long and previous_ma_medium <= previous_ma_long)

    # Price is above long-term MA
    price_above_long_ma = current_price > latest_ma_long

    # MACD is above zero (confirms bullish momentum in the long-term context)
    macd_above_zero = latest_macd > 0

    # OBV showing a long-term upward trend (could check MA of OBV, but for simplicity, just confirming latest uptick here)
    obv_confirm_long_term = latest_obv > previous_obv # Simple confirmation, more advanced checks possible

    # ADX confirming a strong long-term bullish trend (ADX > 25-30) AND +DI is clearly above -DI
    adx_strong_bullish_trend = (latest_adx > 25 and latest_adx_pos > latest_adx_neg * 1.1) # Example: +DI 10% higher than -DI

    # Example Long-Term Buy Logic (more specific combination):
    # Require long-term trend to be bullish (or just turning bullish)
    # AND Momentum confirmation (MACD above zero)
    # AND Price confirms trend (above long MA)
    # AND Volume confirms
    # AND ADX confirms strong trend direction
    long_term_buy_signal = (long_term_uptrend or golden_cross) and \
                           macd_above_zero and \
                           price_above_long_ma and \
                           obv_confirm_long_term and \
                           adx_strong_bullish_trend


    # --- Targets and Stop Loss (Using ATR) ---
    # ATR based targets/stops can be used for both, adjust multipliers as needed

    short_term_target = current_price + (atr_value * target_multiplier)
    short_term_stop_loss = current_price - (atr_value * stop_multiplier)

    long_term_target = current_price + (atr_value * target_multiplier * 3) # Example: 3x short-term distance
    long_term_stop_loss = current_price - (atr_value * stop_multiplier * 2) # Example: 2x short-term stop distance


    # --- Print Analysis ---
    print(f"\nAnalyzing {stock_symbol} for different horizons:")
    print(f"📈 Current Price: ${current_price:.2f}")
    print(f"📊 RSI ({short_rsi_period}): {latest_rsi:.2f}")
    print(f"📉 MAs: Short({short_ma}): {latest_ma_short:.2f}, Medium({medium_ma}): {latest_ma_medium:.2f}, Long({long_ma}): {latest_ma_long:.2f}")
    print(f"📈 MACD: {latest_macd:.2f}, Signal: {latest_macd_signal:.2f}, Diff: {latest_macd_diff:.2f}")
    print(f"📊 BB Lower: {latest_bb_lower:.2f}, BB Upper: {latest_bb_upper:.2f}")
    print(f"📊 Stoch K({stoch_k_period}): {latest_stoch_k:.2f}, Stoch D({stoch_d_period}): {latest_stoch_d:.2f}")
    # OBV value itself is less meaningful, trend is key, but print for completeness
    print(f"📊 OBV: {latest_obv:.2f} (Previous: {previous_obv:.2f})")
    print(f"📊 ADX ({adx_period}): {latest_adx:.2f}, +DI: {latest_adx_pos:.2f}, -DI: {latest_adx_neg:.2f}")
    print(f"📏 ATR ({atr_period}): {atr_value:.2f}")

    print("\n--- Potential Signals ---")

    # Print status of individual potential signals
    if short_rsi_bullish_cross: print("✔ Short-term RSI bullish cross detected.")
    if price_cross_short_ma: print("✔ Price crossed above short-term MA.")
    if short_ma_cross: print("✔ Short MA crossed above Medium MA.")
    if short_macd_bullish_cross: print("✔ Short-term MACD bullish cross detected.")
    if short_bb_bounce: print("✔ Price bounced from Lower Bollinger Band.")
    if stoch_bullish_cross_oversold: print("✔ Stochastic bullish cross from oversold detected.")
    if obv_uptick: print("✔ OBV shows positive volume flow.")
    if adx_confirm_bullish_momentum: print("✔ ADX indicates bullish directional movement.")

    if long_term_uptrend: print("✔ Medium MA is above Long MA (Long-term uptrend).")
    if golden_cross: print("✔ Golden Cross (Medium MA above Long MA) detected.")
    if price_above_long_ma: print("✔ Price is above Long MA.")
    if macd_above_zero: print("✔ MACD is above zero (Bullish momentum).")
    if obv_confirm_long_term: print("✔ OBV confirms long-term volume flow.") # Reusing uptick for simplicity
    if adx_strong_bullish_trend: print("✔ ADX confirms strong long-term bullish trend.")


    # --- Final Buy Signal Check by Horizon ---
    print("\n--- Trading Horizon Signals ---")

    if short_term_buy_signal:
        print("✅ Potential SHORT-TERM BUY Signal (Advanced Combination)")
        print(f"🎯 Suggested Short-Term Target: ${short_term_target:.2f}")
        print(f"🛑 Suggested Short-Term Stop Loss: ${short_term_stop_loss:.2f}")
    else:
        print("❌ No short-term buy signal right now based on advanced criteria.")

    if long_term_buy_signal:
        print("✅ Potential LONG-TERM BUY Signal (Advanced Combination)")
        print(f"🎯 Suggested Long-Term Target: ${long_term_target:.2f}")
        print(f"🛑 Suggested Long-Term Stop Loss: ${long_term_stop_loss:.2f}")
    else:
        print("❌ No long-term buy signal right now based on advanced criteria.")


# Example usage with advanced indicators and horizon-based signals
check_buy_signal_advanced("PLTR") # Example 1
# check_buy_signal_advanced("MSFT") # Example 2
# check_buy_signal_advanced("PLTR") # Example 3
# check_buy_signal_advanced("SPY") # Example 4 (ETF)
```

```
import yfinance as yf
import pandas as pd
import ta

# Added parameters for new indicators and restored target/stop multipliers
def check_buy_signal_super_advanced(stock_symbol, short_rsi_period=14, short_ma=20, medium_ma=50, long_ma=200, atr_period=14, bb_window=20, bb_dev=2, macd_fast=12, macd_slow=26, macd_sign=9, stoch_k_period=14, stoch_d_period=3, cci_period=20, psar_step=0.02, psar_max_step=0.2, target_multiplier=2, stop_multiplier=1):
    """
    Analyzes a stock for potential buy signals using multiple technical indicators.

    Returns:
        dict: A dictionary containing analysis results including signals and price levels,
              or None if analysis failed.
    """
    # Fetch enough data for all indicators
    try:
        df = yf.download(stock_symbol, period="5y", progress=False) # Added progress=False for cleaner output
    except Exception as e:
        print(f"❌ Failed to download data for {stock_symbol}: {e}")
        return None # Return None if data download fails

    # Drop any missing data
    df.dropna(inplace=True)

    # --- Calculate Indicators ---

    # Handle cases where not enough data is available after dropna for indicator periods
    required_data_points = max(short_ma, medium_ma, long_ma, bb_window, atr_period, macd_slow, stoch_k_period, stoch_d_period, cci_period)
    if len(df) < required_data_points:
         print(f"\nAnalyzing {stock_symbol}:")
         print(f"❌ Not enough historical data ({len(df)} data points) to calculate all indicators after dropping NaNs. Need at least {required_data_points} data points.")
         return None # Return None if not enough data

    try:
        # Calculate RSI
        rsi_indicator = ta.momentum.RSIIndicator(df['Close'].squeeze(), window=short_rsi_period)
        df['RSI'] = rsi_indicator.rsi()

        # Moving Averages
        df['MA_Short'] = df['Close'].rolling(window=short_ma).mean()
        df['MA_Medium'] = df['Close'].rolling(window=medium_ma).mean()
        df['MA_Long'] = df['Close'].rolling(window=long_ma).mean()

        # ATR
        atr_indicator = ta.volatility.AverageTrueRange(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=atr_period)
        df['ATR'] = atr_indicator.average_true_range()

        # MACD
        macd_indicator = ta.trend.MACD(close=df['Close'].squeeze(), window_slow=macd_slow, window_fast=macd_fast, window_sign=macd_sign)
        df['MACD'] = macd_indicator.macd()
        df['MACD_Signal'] = macd_indicator.macd_signal()
        df['MACD_Diff'] = macd_indicator.macd_diff()

        # Bollinger Bands
        bb_indicator = ta.volatility.BollingerBands(close=df['Close'].squeeze(), window=bb_window, window_dev=bb_dev)
        df['BB_Lower'] = bb_indicator.bollinger_lband()
        df['BB_Upper'] = bb_indicator.bollinger_hband()
        df['BB_MA'] = bb_indicator.bollinger_mavg()

        # Stochastic Oscillator
        stoch_indicator = ta.momentum.StochasticOscillator(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=stoch_k_period, smooth_window=stoch_d_period)
        df['STOCH_K'] = stoch_indicator.stoch()
        df['STOCH_D'] = stoch_indicator.stoch_signal()

        # On-Balance Volume (OBV)
        obv_indicator = ta.volume.OnBalanceVolumeIndicator(close=df['Close'].squeeze(), volume=df['Volume'].squeeze())
        df['OBV'] = obv_indicator.on_balance_volume()

        # Commodity Channel Index (CCI)
        cci_indicator = ta.trend.CCIIndicator(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), window=cci_period)
        df['CCI'] = cci_indicator.cci()

        # Parabolic SAR
        psar_indicator = ta.trend.PSARIndicator(high=df['High'].squeeze(), low=df['Low'].squeeze(), close=df['Close'].squeeze(), step=psar_step, max_step=psar_max_step)
        df['PSAR'] = psar_indicator.psar()

        # Daily Volume Weighted Average Price (VWAP) - Simple daily calculation
        df['Typical_Price'] = (df['High'] + df['Low'] + df['Close']) / 3
        df['TP_Volume'] = df['Typical_Price'] * df['Volume'].squeeze()
        df['Cumulative_TP_Volume'] = df['TP_Volume'].cumsum()
        df['Cumulative_Volume'] = df['Volume'].cumsum()
        df['VWAP'] = df['Cumulative_TP_Volume'] / df['Cumulative_Volume']
        df = df.drop(columns=['Typical_Price', 'TP_Volume', 'Cumulative_TP_Volume', 'Cumulative_Volume'])

    except Exception as e:
         print(f"\nAnalyzing {stock_symbol}:")
         print(f"❌ Error calculating indicators: {e}")
         return None # Return None if indicator calculation fails


    # Drop any new NaNs after indicators
    df.dropna(inplace=True)

    # Check if there is enough data for comparisons after dropping NaNs
    if len(df) < 2:
        print(f"\nAnalyzing {stock_symbol}:")
        print("❌ Not enough historical data for comparison (need at least 2 rows after indicator calculation).")
        return None # Return None if not enough data for comparisons


    # Use the most recent row and the previous row
    latest = df.iloc[-1]
    previous = df.iloc[-2]


    # Extract scalar values using .item() for latest data
    try:
        current_price = latest['Close'].item()
        atr_value = latest['ATR'].item()
        latest_rsi = latest['RSI'].item()
        latest_ma_short = latest['MA_Short'].item()
        latest_ma_medium = latest['MA_Medium'].item()
        latest_ma_long = latest['MA_Long'].item()
        latest_macd = latest['MACD'].item()
        latest_macd_signal = latest['MACD_Signal'].item()
        latest_macd_diff = latest['MACD_Diff'].item()
        latest_bb_lower = latest['BB_Lower'].item()
        latest_bb_upper = latest['BB_Upper'].item()
        latest_bb_ma = latest['BB_MA'].item()
        latest_stoch_k = latest['STOCH_K'].item()
        latest_stoch_d = latest['STOCH_D'].item()
        latest_obv = latest['OBV'].item()
        latest_cci = latest['CCI'].item()
        latest_psar = latest['PSAR'].item()
        latest_vwap = latest['VWAP'].item()

        # Extract scalar values using .item() for previous day data (for crossover/change checks)
        previous_rsi = previous['RSI'].item()
        previous_ma_short = previous['MA_Short'].item()
        previous_ma_medium = previous['MA_Medium'].item()
        previous_ma_long = previous['MA_Long'].item()
        previous_macd = previous['MACD'].item()
        previous_macd_signal = previous['MACD_Signal'].item()
        previous_macd_diff = previous['MACD_Diff'].item()
        previous_close = previous['Close'].item()
        previous_stoch_k = previous['STOCH_K'].item()
        previous_stoch_d = previous['STOCH_D'].item()
        previous_obv = previous['OBV'].item()
        previous_cci = previous['CCI'].item()
        previous_psar = previous['PSAR'].item()
        previous_vwap = previous['VWAP'].item()

    except Exception as e:
         print(f"\nAnalyzing {stock_symbol}:")
         print(f"❌ Error extracting latest/previous data: {e}")
         return None # Return None if extraction fails


    # --- Define Conditions for Short-Term Buy Signal ---

    short_rsi_bullish_cross = (latest_rsi > 30 and previous_rsi <= 30) or \
                              (latest_rsi > 50 and previous_rsi <= 50)

    price_cross_short_ma = (current_price > latest_ma_short and previous_close <= latest_ma_short)

    short_ma_cross = (latest_ma_short > latest_ma_medium and previous_ma_short <= previous_ma_medium)

    short_macd_bullish_cross = (latest_macd > latest_macd_signal and previous_macd <= previous_macd_signal)

    short_bb_bounce = (current_price > latest_bb_lower and previous_close <= latest_bb_lower)

    stoch_bullish_cross_oversold = (latest_stoch_k > latest_stoch_d and previous_stoch_k <= previous_stoch_d and latest_stoch_k < 20)

    obv_uptick = latest_obv > previous_obv

    cci_bullish_signal = latest_cci > -100 and previous_cci <= -100

    psar_bullish_cross = current_price > latest_psar and previous_close <= previous_psar

    price_above_vwap = current_price > latest_vwap

    # Example Short-Term Buy Logic:
    short_term_buy_signal = (stoch_bullish_cross_oversold or short_macd_bullish_cross or cci_bullish_signal) and \
                            (price_cross_short_ma or short_bb_bounce or psar_bullish_cross or price_above_vwap) and \
                            obv_uptick


    # --- Define Conditions for Long-Term Buy Signal ---

    long_term_uptrend = latest_ma_medium > latest_ma_long

    golden_cross = (latest_ma_medium > latest_ma_long and previous_ma_medium <= previous_ma_long)

    price_above_long_ma = current_price > latest_ma_long

    macd_above_zero = latest_macd > 0

    obv_confirm_long_term = latest_obv > previous_obv # Simple confirmation

    cci_above_zero = latest_cci > 0

    price_above_psar_trend = current_price > latest_psar

    # Example Long-Term Buy Logic:
    long_term_buy_signal = (long_term_uptrend or golden_cross) and \
                           (macd_above_zero or cci_above_zero) and \
                           (price_above_long_ma or price_above_psar_trend) and \
                           obv_confirm_long_term


    # --- Targets and Stop Loss (Using ATR) ---
    short_term_target = current_price + (atr_value * target_multiplier)
    short_term_stop_loss = current_price - (atr_value * stop_multiplier)

    long_term_target = current_price + (atr_value * target_multiplier * 3)
    long_term_stop_loss = current_price - (atr_value * stop_multiplier * 2)


    # --- Return Analysis Results ---
    result = {
        "symbol": stock_symbol,
        "current_price": current_price,
        "short_term_signal": False,
        "long_term_signal": False,
        "short_term_target": None,
        "short_term_stop_loss": None,
        "long_term_target": None,
        "long_term_stop_loss": None,
        "analysis_successful": True # Flag to indicate if analysis completed without error
    }

    if short_term_buy_signal:
        result["short_term_signal"] = True
        result["short_term_target"] = short_term_target
        result["short_term_stop_loss"] = short_term_stop_loss

    if long_term_buy_signal:
        result["long_term_signal"] = True
        result["long_term_target"] = long_term_target
        result["long_term_stop_loss"] = long_term_stop_loss

    # You might want to add more details to the result dictionary, like indicator values

    return result


# --- Example Usage to find top stocks ---

# Define a list of stocks to analyze (replace with your desired list)
stock_list = ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA", "NVDA", "META", "BRK-B", "JPM", "V", "MA", "UNH", "LLY", "XOM", "PG", "JNJ", "HD", "PFE", "DIS", "CRM", "PLTR", "CRDO", "SMCI", "AMD"]

analysis_results = []

print("Starting stock analysis...")
for stock in stock_list:
    print(f"\nAnalyzing {stock}...")
    result = check_buy_signal_super_advanced(stock)
    if result is not None and result["analysis_successful"]:
        analysis_results.append(result)
    # Add a small delay to avoid hitting API limits too hard (optional)
    # import time
    # time.sleep(1)

print("\n--- Analysis Summary ---")

short_term_buys = [r for r in analysis_results if r["short_term_signal"]]
long_term_buys = [r for r in analysis_results if r["long_term_signal"]]

print(f"\nPotential Short-Term Buy Signals ({len(short_term_buys)} stocks):")
if short_term_buys:
    # You could sort these based on criteria if needed
    for stock_info in short_term_buys:
        print(f"- {stock_info['symbol']}: Current Price ${stock_info['current_price']:.2f}, Target ${stock_info['short_term_target']:.2f}, Stop Loss ${stock_info['short_term_stop_loss']:.2f}")
else:
    print("None")


print(f"\nPotential Long-Term Buy Signals ({len(long_term_buys)} stocks):")
if long_term_buys:
     # You could sort these based on criteria if needed
    for stock_info in long_term_buys:
        print(f"- {stock_info['symbol']}: Current Price ${stock_info['current_price']:.2f}, Target ${stock_info['long_term_target']:.2f}, Stop Loss ${stock_info['long_term_stop_loss']:.2f}")
else:
    print("None")


# To get "Top 10", you'd need a way to rank them.
# For a simple "top stocks with a buy signal", you can just list them.
# If you wanted a ranked "Top 10", you'd need to define a scoring system
# based on how many conditions were met, how strong the signals are, etc.,
# and then sort the `analysis_results` list before selecting the top 10.

# Example: Simple list of stocks with ANY buy signal
all_buy_signals = short_term_buys + [r for r in long_term_buys if r not in short_term_buys] # Avoid duplicates if a stock has both
print(f"\nStocks with Any Potential Buy Signal ({len(all_buy_signals)} stocks):")
if all_buy_signals:
    # To get "Top 10", you would need a sorting key here.
    # For now, we just print the list.
    for stock_info in all_buy_signals[:10]: # Print up to the first 10 found
         signal_type = ""
         if stock_info["short_term_signal"] and stock_info["long_term_signal"]:
             signal_type = " (Short & Long Term)"
         elif stock_info["short_term_signal"]:
             signal_type = " (Short Term)"
         elif stock_info["long_term_signal"]:
              signal_type = " (Long Term)"
         print(f"- {stock_info['symbol']}{signal_type}: Price ${stock_info['current_price']:.2f}")
else:
    print("None")
```
